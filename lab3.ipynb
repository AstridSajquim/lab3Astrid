```python
# Definir los arrays
a = [1, 4, 2, 7]
b = [1, 3, 2, 9]

# Calcular la diferencia entre los elementos correspondientes de los arrays
diferencias = [a[i] - b[i] for i in range(len(a))]

# Calcular la media de las diferencias
media_diferencias = sum(diferencias) / len(diferencias)

print("La media aritmética de la diferencia (a-b) es:", media_diferencias)
```

```python
import numpy as np

# Definir el array con los valores de coordenadas
coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

# Copiar y transponer el array
coordenadas_transpuesta = coordenadas.T.copy()

# Agregar las nuevas coordenadas al array transpuesto
nuevas_coordenadas = np.array([[10.8, 35.8], [3.0, 12.0]])
coordenadas_transpuesta = np.hstack([coordenadas_transpuesta, nuevas_coordenadas])

# Transponer el array de nuevo
coordenadas_final = coordenadas_transpuesta.T

# Mostrar las dimensiones del nuevo array
print("Dimensiones del nuevo array:", coordenadas_final.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("nucleospoblacion.csv")

# Ordenar el DataFrame por población de menor a mayor
df_sorted = df.sort_values(by='Poblacion')

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['Municipio'], df_sorted['Poblacion'], color='magenta')
plt.xlabel('Ciudad')
plt.ylabel('Población')
plt.title('Población de cada ciudad (de menor a mayor)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()



```

```python
# Encontrar la segunda ciudad más poblada
segunda_ciudad = df_sorted.iloc[1]['Municipio']
print("La segunda ciudad más poblada es:", segunda_ciudad, df_sorted.iloc[1]['Poblacion'])
```

```python
# Encontrar la posición de Granada en el ranking de las ciudades más pobladas
posicion_granada = df_sorted[df_sorted['Municipio'] == 'Granada'].index[0] + 1
print("Granada ocupa la posición", posicion_granada, "en el ranking de las ciudades más pobladas.")
```

# 5


```python
# Filtrar los municipios de la provincia de Granada
municipios_granada = df[df['Provincia'] == 'Granada']

# Calcular las medidas estadísticas de la población
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
valor_maximo_poblacion = municipios_granada['Poblacion'].max()
valor_minimo_poblacion = municipios_granada['Poblacion'].min()

# Imprimir los resultados
print("Media de población:", media_poblacion)
print("Mediana de población:", mediana_poblacion)
print("Desviación estándar de población:", desviacion_estandar_poblacion)
print("Valor máximo de población:", valor_maximo_poblacion)
print("Valor mínimo de población:", valor_minimo_poblacion)
```

```python
import matplotlib.pyplot as plt

# Obtener la lista de provincias en el DataFrame
provincias = df['Provincia'].unique()

# Configurar el tamaño de la figura y la disposición de los subgráficos
num_provincias = len(provincias)
fig, axs = plt.subplots(num_provincias, figsize=(10, 5*num_provincias))

# Iterar sobre cada provincia y crear un gráfico
for i, provincia in enumerate(provincias):
    # Seleccionar los datos de la provincia actual
    datos_provincia = df[df['Provincia'] == provincia]
    municipios = datos_provincia['Municipio']
    poblaciones = datos_provincia['Poblacion']
    # Crear el gráfico en la posición adecuada
    axs[i].barh(municipios, poblaciones, color='magenta')
    axs[i].set_title(provincia)
    axs[i].set_xlabel('Población')
    axs[i].set_ylabel('Municipio')

# Ajustar el diseño de la figura
plt.tight_layout()
plt.show()

```

```python
import random

# Seleccionar cincuenta municipios al azar sin repeticiones
municipios_aleatorios = random.sample(list(df['Municipio'].unique()), 50)

# Filtrar los datos de los municipios seleccionados
datos_seleccionados = df[df['Municipio'].isin(municipios_aleatorios)]

# Calcular el promedio y la desviación estándar de la población de los municipios seleccionados
promedio_poblacion = datos_seleccionados['Poblacion'].mean()
desviacion_estandar_poblacion = datos_seleccionados['Poblacion'].std()

# Imprimir los resultados
print("Promedio de la población de los 50 municipios seleccionados:", promedio_poblacion)
print("Desviación estándar de la población de los 50 municipios seleccionados:", desviacion_estandar_poblacion)
```

```python
import numpy as np

# Coordenadas de Madrid
madrid_x = -3.7038
madrid_y = 40.4168

# Calcular la distancia entre Madrid y cada municipio
df['Distancia_a_Madrid'] = np.sqrt((df['X'] - madrid_x)**2 + (df['Y'] - madrid_y)**2)

# Excluir Madrid de los cálculos (distancia cero)
df_sin_madrid = df[df['Municipio'] != 'Madrid']

# Encontrar el municipio más cercano a Madrid
municipio_mas_cercano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]
distancia_cercano = municipio_mas_cercano['Distancia_a_Madrid']
nombre_cercano = municipio_mas_cercano['Municipio']

# Encontrar el municipio más lejano a Madrid
municipio_mas_lejano = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]
distancia_lejano = municipio_mas_lejano['Distancia_a_Madrid']
nombre_lejano = municipio_mas_lejano['Municipio']

print("El municipio más cercano a Madrid es:", nombre_cercano)
print("Distancia:", distancia_cercano, "kilómetros")

print("El municipio más lejano a Madrid es:", nombre_lejano)
print("Distancia:", distancia_lejano, "kilómetros")

```
